*Qu es .gitignore?*

Es un archivo de texto plano que le dice a Git qu archivos o carpetas debe ignorar (es decir, no rastrear ni subir al repositorio).

Ejemplo: .gitignore

```python
# __pycache__ y archivos .pyc (compilados)
__pycache__/
*.pyc

# Entorno virtual de Python
venv/

# Archivos de Jupyter temporales
.ipynb_checkpoints/

# Configuracin de VS Code
.vscode/

# Archivos de entorno o secretos
.env
secrets.json

# Logs
*.log
```

*Se debe subir el .gitignore?*

S.

La primera vez lo agregas, haces commit y lo subes para que todos en el equipo (y Jenkins) lo usen igual.

*Cundo entra en accin el .gitignore?*

El .gitignore entra en juego desde que haces git add, no solo en el git push.

Si haces git add archivo.py, solo ese archivo se agrega, el .gitignore no influye.

Si haces git add <directorio> o git add ., Git ignora todo lo que est listado en .gitignore.

Si modificas o creas un .gitignore, debes hacer git add .gitignore para que el cambio quede registrado en Git.

*Ramas remotas*

* Ramas locales (ej: main)

* Ramas remotas (remotes/origin/rama-de-casa, remotes/origin/rama-de-oficina)



1. *Trababjar sobre una Rama existente*

Si ya existe en remoto, solo tienes que "trartela" y cambiarte:

```python
git checkout -b rama-de-casa origin/rama-de-casa
```

Esto crea una rama local conectada a la remota rama-de-casa.

2. *Trabajar sobre una rama existente : forma 2*

Si quieres una rama que no existe todava (ej: rama-de-oficina), estando en tu repo local haces:

```python
git checkout -b rama-de-oficina
```

(eso crea la rama en local).

Luego, si quieres mandarla al remoto:

```python
git push -u origin rama-de-oficina
```

El -u sirve para que quede trackeada (es decir, que tu rama local sepa de cul remota bajar cambios).

*actualizamos las ramas remotas en donde deseamos trabajar*

```python
git fetch origin
```

```python
git fetch origin --prune
```

- git fetch origin:
Esto descarga los cambios ms recientes del repositorio remoto llamado origin (por lo general, el repositorio principal en GitHub o similar). No modifica tu rama actual, solo actualiza las referencias remotas.
- --prune:
Esta opcin elimina las referencias remotas que ya no existen en el servidor. Por ejemplo, si alguien borr una rama en el repositorio remoto, --prune har que desaparezca tambin de tu lista local de ramas remotas (remotes/origin/xyz).
